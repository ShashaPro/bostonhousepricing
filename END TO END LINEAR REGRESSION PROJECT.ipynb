# R square and adjusted R square

Formula  <br />

<b>R^2 = 1 - SSR/SST</b> <br />

R^2 = coefficient of determinent SSR=sum of squares of residuals SST = total sum of squares



```python
from sklearn.metrics import r2_score
score = r2_score(y_test, reg_pred)
print(score)
```

<h1><b>Adjusted R2 = 1 - [(1 - R2)*(n - 1) / (n - k - 1)]</b></h1> <br />

<p>where : </p>
<p>R2: The R2 of the model n: The number of observation k: The number of predictor variable</p>

```python
## display adjusted R-squared
1 - (1-score)*(len(y_test)-1) / (len(y_test) - X_test.shape[1] - 1)
```
